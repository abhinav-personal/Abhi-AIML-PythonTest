```python
def func1 (**a):
  print(a)
  print(type(a))
  print(a.keys())
  print(a["city"])

func1(name = "Abhinav", age = 41, city = "Ranchi")
dict1 = {"name" : "Abhinav", "age" : 41, "city" : "Ranchi"}
print(dict1["city"])
```

```python
#conbining Positional arguments and Keyword arguments
def func1 (*a, **b):
  print(a)
  print(b)

func1("Abhinav", 41, city = "Ranchi")
```

```python
#difining different argument
def func1 (name, age, *pos_arg, city, state = "Jharkhand", **key_arg):
  print(name)
  print(age)
  print(pos_arg)
  print(city)
  print(state)
  print(key_arg)

func1("Abhinav", 41, "Graduate", "BDCOE", city = "Ranchi", locality = "Bariatu", Pin = 834009)
```
