```python
vl1 = [321255, "Abhinav", "NTT Data", "US"]
vd1 = {1:"Abhi", 2:"Srivastava"}
vd2 = {3:"Abhinav", 4:"Sriva"}
print(vd1.keys())
print(vd1.values())
vd3 = vd1.copy()
vd1.update(vd2)
print(vd3)
print(vd1)
vl1.append(vd1)
print(vl1)
print(vd1)
```

```python
# iterate a dict for key and value
vd1 = {1:"Abhi", 2:"Srivastava"}
print (vd1.keys())
for i in vd1:
  print(i)

for j in vd1.values():
  print(j)

for k in vd1.items():
  print(k)
print("\n")

for a,b in vd1.items():
  print(a)
  print(b)
```

```python
k=[1,2,3,4,5]
val=[10,20,30,40,50]
for i in val:
  d1 = dict.fromkeys(k,i)
print(d1)

d2 = dict(zip(k,val))
print(d2)
```

```python
s= " I live in Ranchi I find Ranchi a beautiful city"
s1= s.split()
print(s1)
```

```python
word_counts = {}
for word in s1:
    print(word)
    print(word_counts.get(word, 0) + 1)
    word_counts[word] = word_counts.get(word, 0)+1
d1 = word_counts
print(d1)
```
