```python
vl1 = [321255, "Abhinav", "NTT Data", "US"]
vd1 = {1:"Abhi", 2:"Srivastava"}
vd2 = {3:"Abhinav", 4:"Sriva"}
print(vd1.keys())
print(vd1.values())
vd3 = vd1.copy()
vd1.update(vd2)
print(vd3)
print(vd1)
vl1.append(vd1)
print(vl1)
print(vd1)
```

```python
# iterate a dict for key and value
vd1 = {1:"Abhi", 2:"Srivastava"}
print (vd1.keys())
for i in vd1:
  print(i)

for j in vd1.values():
  print(j)

for k in vd1.items():
  print(k)
print("\n")

for a,b in vd1.items():
  print(a)
  print(b)
```
