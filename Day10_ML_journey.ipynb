```python
#Mini Project
def student_database():
  #create a empty database
  student_databse = {}
  return student_databse

def add_student_data(database):
  sid = input("Enter Student ID: ")
  sname = input("Enter Student Name: ")
  smajor = input("Enter Student Major: ")
  database[sid] = {"name": sname, "major": smajor}
  print("Student " + sname + " added successfully.")


def find_student_info(database):
  #get studentID from the user to find
  sid = input("Enter Student ID to find: ")
  if sid in database:
    student_data = database.get(sid)
    return student_data
  else:
    return "student not found"

def delete_student_data(database):
  #get studentID from the user to delete
  sid = input("Enter Student ID to delete: ")
  if sid in database:
    del database[sid]
    return "Student deleted successfully"
  else:
    return "Student not found"

def update_student_data(database):
  #get studentID to update
  sid = input("Enter Student ID to update: ")
  if sid in database:
    #get new student data
    new_name = input("Enter new name: ")
    new_major = input("Enter new major: ")
    database[sid] = {"name": new_name, "major": new_major}
    return "Student updated successfully"
  else:
    return "Student not found"

def display_student_data(database):
  print("Student Database:")
  if database:
    print(database)
  else:
    print("No students in the database.")

#main Program
student_database = student_database()
while True:
  print("\nStudent Database Menu:")
  print("1. Add Student Data")
  print("2. Find Student Info")
  print("3. Delete Student Data")
  print("4. Update Student Data")
  print("5. Display Student Data")
  print("6. Exit")

  choice = input("Enter your choice (1-6): ")
  if choice == "1":
    add_student_data(student_database)

  elif choice == "2":
    student_data = find_student_info(student_database)
    print(student_data)

  elif choice == "3":
    result = delete_student_data(student_database)
    print(result)

  elif choice == "4":
    result = update_student_data(student_database)
    print(result)

  elif choice == "5":
    display_student_data(student_database)

  elif choice == "6":
    print("Exiting the program.")
    break

  else:
    print("Invalid choice. Please enter a number between 1 and 6.")


```
