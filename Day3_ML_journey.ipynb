```python
""" variable annotations is to let people know how this variable should be treated.
not enforcing the datatype"""
EmpID: int = "NTT"
print(EmpID)
print(type(EmpID))
print(__annotations__['EmpID'])
```

```python
#tuple is also a replacement of array but Immutable
T1= (10,23.4,"NTT")
print(T1[-1])
```

```python
# Dictionary is also a replacement of array but in key value pair.
# This also gives the option to create the index value of our own

D1 = {'a':1, 'b':2, 10:"test"}
print(D1['a'])
```

```python
# Set is also a replacement of array but in unordered format.

S1 = {1, 2, "test"}
print(S1)
```
