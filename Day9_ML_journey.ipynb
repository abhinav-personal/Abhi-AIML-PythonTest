```python
import copy
l1 = [10,20,30,40]
l2 = l1
print(l1)
print(l2)
print(id(l1))
print(id(l2))

l2[2] = "Test"

print(l1)
print(l2)
print(id(l1))
print(id(l2))

l3 = copy.copy(l1)
print(l1)
print(l3)
print(id(l3))
l3[1] = "NTT"
print(l1)
print(l3)
print(id(l3))
```

```python
#shallow copy
import copy
l1 = [[1,2,3], [4,5], ["Test", "Abhinav"]]
l2 = copy.copy(l1)
print("Id of list 1: ",  id(l1))
print("Id of List 2: ", id(l2))
for i in l1:
  print("Id of ", i , "element of List 1 is: ", id(i))

for k in l2:
  print("Id of ", k , "element of List 2 is: ", id(k))

l1.append(34)
print(l1)
print(l2)
```

```python
#deep copy
#shallow copy
import copy
l1 = [[1,2,3], [4,5], ["Test", "Abhinav"]]
l2 = copy.deepcopy(l1)
print("Id of list 1: ",  id(l1))
print("Id of List 2: ", id(l2))
for i in l1:
  print("Id of ", i , "element of List 1 is: ", id(i))

for k in l2:
  print("Id of ", k , "element of List 2 is: ", id(k))

l1.append(34)
print(l1)
print(l2)
```
