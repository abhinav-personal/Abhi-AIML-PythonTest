```python
#class variable
class test():
  name = "Abhinav"

  def __init__(self,id, city):
    self.name = test.name
    self.id = id
    self.city = city

  def print_details(self):
    print("name: ", self.name, "and id: ", str(self.id), " and lives in: ", self.city)

obj1 = test(1, "Ranchi")
obj2 = test(2, "Lucknow")
obj1.print_details()
obj2.print_details()
print(obj1.id)
print(test.name)
```

```python
# Class Method
class emp_details():
  name = "Abhinav"

  def __init__(self,id, city):
    self.name = emp_details.name
    self.id = id
    self.city = city

  @classmethod
  def display_name(cls):
    print(cls.name)

  def print_details(self):
    print("name: ", self.name, "and id: ", str(self.id), " and lives in: ", self.city)

obj1 = emp_details(1, "Ranchi")
obj2 = emp_details(2, "Lucknow")
obj1.print_details()
obj2.print_details()

emp_details.display_name()

print(obj1.id)
print(emp_details.name)
```
