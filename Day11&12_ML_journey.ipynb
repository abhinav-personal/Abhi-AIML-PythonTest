```python
import traceback
import sys

try:
  val1_str = input("Enter the first value: ")
  val2_str = input("Enter the second value: ")

  try:
    val1 = int(val1_str)
    val2 = int(val2_str)

    div1 = val1/val2
    print(div1)
  except ValueError:
    print("Invalid input. Please enter valid integers.")
    exc_type, exc_value, exc_traceback = sys.exc_info()
    if exc_traceback:
      fname, lineno, funname, line = traceback.extract_tb(exc_traceback)[-1]
      print(f"Error occurred at file: {fname}, line: {lineno}, in function: {funname}")
  except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
    exc_type, exc_value, exc_traceback = sys.exc_info()
    if exc_traceback:
      fname, lineno, funname, line = traceback.extract_tb(exc_traceback)[-1]
      print(f"Error occurred at file: {fname}, line: {lineno}, in function: {funname}")
  except Exception as e:
    print(f"An unexpected error occurred: {e}")
    exc_type, exc_value, exc_traceback = sys.exc_info()
    if exc_traceback:
      fname, lineno, funname, line = traceback.extract_tb(exc_traceback)[-1]
      print(f"Error occurred at file: {fname}, line: {lineno}, in function: {funname}")
finally:
  print("Finished the operation")
```

```python
#ATM Project
class invalidPin(Exception):
  pass

class invalidBalance(Exception):
  pass

attempt = 1
def withdraw(saved_pin, saved_balance):
  global attempt
  try:
    pin = int(input("Enter the pin: "))
    if pin != saved_pin:
      raise invalidPin("Invalid Pin")
    else:
      amount = int(input("Enter the amount: "))
      if amount > saved_balance:
        raise invalidBalance("Insufficient Balance")

      saved_balance = saved_balance - amount
      print("Remaining Balance: ", saved_balance)
      return saved_balance
  except invalidPin as obj:
    print(obj)
    attempt = attempt + 1
    try_again = input("Do you want to try again, y/n? ")
    if try_again == 'y' and attempt <= 3:
      saved_balance = withdraw(saved_pin, saved_balance)
      return saved_balance
    else:
      print("You have exceeded the number of attempts, your account is blocked for 2 Hrs")
      return "Exceed_Retry"
  except invalidBalance as obj:
    print(obj)
    return saved_balance
  except Exception as obj:
    print(obj)
    return saved_balance


saved_pin = 1234
saved_balance = 100000

while saved_balance > 0:
  saved_balance = withdraw(saved_pin, saved_balance)
  next_tran = input("You want to do another transaction, y/n? ")
  if next_tran == 'n' or saved_balance == "Exceed_Retry":
      break

print("Thanks for banking with us, Have a great day!!!")
```
