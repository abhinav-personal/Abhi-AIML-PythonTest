```python
class MyClassWithArgs:
    def __init__(self, name, *values):
        self.name = name
        self.data = []
        print(f"Initializing {self.name} with values: {values}")
        for val in values:
            self.data.append(val)

    def display_data(self):
        print(f"Object Name: {self.name}")
        print(f"Stored Data: {self.data}")

# Example usage:
obj1 = MyClassWithArgs("First Object", 10, 20, 30)
obj1.display_data()

obj2 = MyClassWithArgs("Second Object", "apple", "banana")
obj2.display_data()

obj3 = MyClassWithArgs("Third Object") # No additional values
obj3.display_data()
```

```python
#Calculate employee bonus
class empBonum():

  def __init__(self, empid, empname, empsal):
    self.empid = empid
    self.empname = empname
    self.empsal = empsal

  def calcBonum(self):
    self.empsal = self.empsal + self.empsal * 0.05
    return self.empsal

emp1 = empBonum(1234, "Abhinav", 20000)
final_salary = emp1.calcBonum()
print("The final salary of emp:", emp1.empname, "having empID:", str(emp1.empid), "is", final_salary)
```
