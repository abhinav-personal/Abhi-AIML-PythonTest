```python
class calculator():

  def __init__(self, n1, n2): #special method to create the constructor
    self.n1 = n1
    self.n2 = n2
    self.result = 0

  def add(self):
    self.result =  self.n1 + self.n2
    return self.result

  def sub(self):
    self.result =  self.n1 - self.n2
    return self.result

  def mul(self):
    self.result =  self.n1 * self.n2
    return self.result

  def div(self):
    self.result =  self.n1 / self.n2
    return self.result

print("start of program")
num1 = int(input("enter first number "))
num2 = int(input("enter second number "))

operation = input("enter operation you want to perform: add/sub/mul/div? ")

if operation == "add":
  obj1 = calculator(num1,num2)
  obj1.add()
  print("result of add is", obj1.result)
elif operation == "sub":
  obj2 = calculator(num1,num2)
  obj2.sub()
  print("result of sub is", obj2.result)
elif operation == "mul":
  obj3 = calculator(num1,num2)
  obj3.mul()
  print("result of mul is", obj3.result)
elif operation == "div":
  obj4 = calculator(num1,num2)
  obj4.div()
  print("result of div is", obj4.result)
else:
  print("invalid operation")

print("end of program")
```
