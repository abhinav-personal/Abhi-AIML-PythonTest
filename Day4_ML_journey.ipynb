```python
# Decision making statements in python
# else - if statement
val1 = input("Enter your val1: ") # input funtion always return the input as String
val1 = int(val1)

val2 = input("Enter your val2: ") # input funtion always return the input as String
val2 = int(val2)

if val1 > val2:
  print("greater")
elif val1 == val2:
  print("equal")
else:
  print("lesser")
```

```python
name = input("Enter a name: ")
name = str(name)
print(name[::-1])
if name == name[::-1]:
  print(name + ": is a palindrome " + "word")
else:
  print(name + ": is not a palindrome")
```

```python
#while loop
id1 = input("Enter your id1: ")
id1 = int(id1)
id2 = input("Enter your id2: ")
id2 = int(id2)
while(id1 < id2):
  print(id1,  " is less than ",  id2)
  id1 = id1 + 1

print("end of while")
```

```python
number = input("Enter a number you want the table of: ")
number = int(number)
i=1
print("Table of", number, ":")
print("-------------")
while(i<=10):
  print(number, "x", i, "=", number*i)
  i = i + 1
```
